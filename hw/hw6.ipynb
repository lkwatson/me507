```python
import math
import numpy as np
import scipy as sp
from scipy import integrate
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 4

```python
def y_decay(t):
    return np.exp(-t) + np.exp(-10*t)
```

```python
t_rng = np.linspace(0, 4, 100)
plt.plot(t_rng, y_decay(t_rng))
plt.title("Decay: poles at -1 and -10")
plt.show()
```

```python
def y_step_resp(t):
    return 2*((1-np.exp(-t)) + (1-np.exp(-10*t)))
```

```python
t_rng = np.linspace(0, 4, 100)
plt.plot(t_rng, y_step_resp(t_rng))
plt.title("Step response: poles at -1 and -10")
plt.show()
```

## Problem 5

```python
def rc_circ(e, t):
    return -0.5*e + 2
```

```python
ts_d2 = np.arange(0.0, 20.0, 0.2)
ts_1 = np.arange(0.0, 20.0, 1)
ts_4 = np.arange(0.0, 20.0, 4)

resp_d2 = integrate.odeint(rc_circ, 0.0, ts_d2)
resp_1 = integrate.odeint(rc_circ, 0.0, ts_1)
resp_4 = integrate.odeint(rc_circ, 0.0, ts_4)

plt.plot(ts_d2, resp_d2, label="Timestep 0.2")
plt.plot(ts_1, resp_1, label="Timestep 1")
plt.plot(ts_4, resp_4, label="Timestep 4")
plt.legend()
plt.show()
```
