```python
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# s(s+1)(s+5) -> s^3 + 6s^2 + 5s
s1 = signal.lti([10], [1, 6, 5, 0])
w, mag, phase = signal.bode(s1)

plt.semilogx(w, mag)
plt.grid()
plt.title("Bode — Magnitude")
plt.show()

plt.semilogx(w, phase)
plt.grid()
plt.title("Bode — Phase")
plt.show()
```
