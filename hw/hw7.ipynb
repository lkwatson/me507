```python
import math
import numpy as np
import scipy as sp
from scipy import integrate
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 3

```python
def sys(state, t, k, m, b):
    e, f = state
    return k*f, (-1/m)*e + (-b/m)*f + (1/m)*1 #1 is the step input
```

```python
ts = np.arange(0.0, 20.0, 0.2)

resp_a = integrate.odeint(sys, (0, 0), ts, (3, 1, 0.6))
resp_b = integrate.odeint(sys, (0, 0), ts, (3, 1, 2))

plt.title("e2 vs time")
plt.plot(ts, resp_a[:,0], label="b=0.6")
plt.plot(ts, resp_b[:,0], label="b=2")
plt.legend()
plt.show()

plt.title("f3 vs time")
plt.plot(ts, resp_a[:,1], label="b=0.6")
plt.plot(ts, resp_b[:,1], label="b=2")
plt.legend()
plt.show()
```
